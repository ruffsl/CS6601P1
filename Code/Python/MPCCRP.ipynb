# Multi-Party Capacity Constrained Route Planning

MPCCRP is a Emergency Route Planning algorithm designed for multiple party coordination. Such applications include Civilian Evacuation and Coordinated with Emergency Response Teams. To start off click import some dependencies including **ccrpTools**, our own custom library with an MPCCRP implementation.  

Behind the scenes, we will also be using:  
**random**  
**heapq**  
**pandas**  
**networkx**  
**matplotlib**  

```python
from ccrpTools import *
```

Here we'll set a few parameters describing the range of possibilities for our later generated capacity constrained network. Will define the maximin possible populations of all nodes for each party: Evacuees, and Responders. We will also set the maximum node and edge capacities along with the range of travel times for any given edge.

```python
popN = {'evac':{'min':2,'max':5},
        'resp':{'min':0,'max':1}}
capN  = {'min':1,'max':5}
capE  = {'min':1,'max':5}
timeE = {'min':1,'max':5}
```

Lets generate a small 5 by 5 two-dimensional capacity constrained Manhattan grid by constructing a MPCCRP object using the parameters we defined above.

```python
mpccrp = MPCCRP(popN, capN, capE, timeE, graph_type='grid_2d', graph_shape=[5, 5])
```

Now lets define an area of the grid an evacuation zone, say the 3 by 3 interior.
From this we will get the sources and destinations for all that affected evacuees.

```python
Sevac, Devac = mpccrp.genSD(source_shape=[[1, 4], [1, 4]])
```

We'll go ahead and make the very center of the grid the disaster node, thus the destination for all responders.  
We'll also make sure that no responders are already on-site of the disaster at the beginning time of the simulation.  
Simply read from the graph to determine their starting positions to be used as their source locations.

```python
Dresp = [(2,2)]
mpccrp.G.node[(4,4)]['pop']['resp'] = {0:0}
Sresp = mpccrp.getPartyN('resp', 0)
```

Now will generate a simple dictionary to describe to our MPCCRP algorithm all the sources and destinations for each party.

```python
SD = {'evac':{'S':Sevac,'D':Devac},
      'resp':{'S':Sresp,'D':Dresp}}
```

Let's go ahead and see what this looks like so far given the starting time of 0.  
While we're at it, let's go ahead check what the total population is too.

```python
print 'Population Before: ', mpccrp.getPop(t=0)
fig = plt.figure(figsize=(10,10))
figG = mpccrp.drawGraph(SD=SD, params='ne', t=0)
plt.show()
```

The figure above is a visual representation of our recently defined capacity constrained network.  
Node numbers represent maximum capacity, denoted by the size of the shaded blue outline and population colored red for evacuation zone and orange for an existing responder.  
The edge numbers represent edge capacity denoted by edge width and travel time denoted by green to yellow for fast to slow.

If we would like to solve for an evacuation route plan  
including the paths for the emergency responders to the disaster node,  
we can simply call our MPCCRP to generate an entire plan from start to finish.

```python
plan, totalTime = mpccrp.applyCCRP(SD, t=0)
```

Let's go ahead and see what the end result looks like

```python
print 'Population After: ', mpccrp.getPop(t=totalTime)
fig = plt.figure(figsize=(10,10))
figG = mpccrp.drawGraph(SD=SD, params='ne', t=totalTime)
plt.show()
```

You'll notice that all evacuees have exited the evacuation zone through the quickest routes available, and thus coagulate around the major exits consisting of wider and faster edges. The emergency responders have also accumulated onto the center disaster node. we can also check to see that the population has remained conservative with everyone accounted for.

Lets see what the first and last evacuation stages in the plan look like, and how much time the total evacuation plan will take

```python
print 'Total Time', totalTime
print 'First stage'
plan[1]
```

Above we can see that one ('Flow' is 1) Evacuee ('Party' is 'evac') located at the source 'S' is assigned to evacuate to destination 'D' using the path 'R' at the 'Start' time at t = 0. The plan also holds additional information about other relevant population and reservation bookkeeping for both source and destination

```python
print 'Last stage'
plan[-1]
```

Above we can see that the emergency responder located at the bottom left of the graph is finally assigned the necessary route to reach the center disaster node

Now let's do a comparison between using CCRP for one party after another  
VS parallelizing the efforts of multiple party planning using MPCCRP.  
This exparement looks at the results after 1000 trials.  
This may take a while...

```python
data = []
N = 1000
graph_shape = [9,9]
ccrp = MPCCRP(popN, capN, capE, timeE, graph_shape=graph_shape, seed=42)
Sevac, Devac = ccrp.genSD(source_shape=[[1, 8], [1, 8]])
Dresp = [(4,4)]
for i in range(N):
    trial = {}
    print "Trial: ", i
    trial["Trial"] = i
    
    seed = random.random()
    mpccrp_new = MPCCRP(popN, capN, capE, timeE, graph_shape=graph_shape, seed=seed)
    mpccrp_new.G.node[(4,4)]['pop']['resp'] = {0:0}
    Sevac_new = list(Sevac)
    Devac_new = list(Devac)
    Sresp_new = mpccrp_new.getPartyN('resp', 0)
    Dresp_new = list(Dresp)
    SD_new = {'evac':{'S':Sevac_new,'D':Devac_new},
              'resp':{'S':Sresp_new,'D':Dresp_new}}
    plan_new, totalTime_new = mpccrp_new.applyCCRP(SD_new, t=0)
    
    print "totalTime_new: ", totalTime_new
    trial["totalTime_new"] = totalTime_new
    
    ccrp_old = MPCCRP(popN, capN, capE, timeE, graph_shape=graph_shape, seed=seed)
    ccrp_old.G.node[(4,4)]['pop']['resp'] = {0:0}
    Sevac_old = list(Sevac)
    Devac_old = list(Devac)
    SD_evac_old = {'evac':{'S':Sevac_old,'D':Devac_old}}
    plan_old, totalTime_evac_old = ccrp_old.applyCCRP(SD_evac_old, t=0)
    Sresp_old = ccrp_old.getPartyN('resp', 0)
    Dresp_old = list(Dresp)
    SD_resp_old = {'resp':{'S':Sresp_old,'D':Dresp_old}}
    plan_old, totalTime_old = ccrp_old.applyCCRP(SD_resp_old, t=totalTime_evac_old)
    
    
    print "totalTime_evac_old: ", totalTime_evac_old
    trial["totalTime_evac_old"] = totalTime_evac_old
    print "totalTime_resp_old: ", totalTime_old - totalTime_evac_old
    trial["totalTime_resp_old"] = totalTime_old - totalTime_evac_old
    print "totalTime_old: ", totalTime_old
    trial["totalTime_old"] = totalTime_old
    
    data.append(trial)
    
df = pd.DataFrame(data)
df = df.set_index('Trial')
```

Ok, now that's done, lets take a look at the dataframe of trials we've generated

Perhaps it would be more useful see the results visualized as a histogram.  
We can see here that on average it does take longer to move both parties that it does for just evacuees

```python
fig = pyplot.figure(figsize=(5,5), dpi=80)
ax = fig.add_subplot(1,1,1)
plt.hold()
plt.hist(df['totalTime_evac_old'], 20, alpha=0.7, color='g', label='Old Evacuation')
plt.hist(df['totalTime_new'], 30, alpha=0.7, color='b', label='New Total')
plt.xlim((10,40))
plt.grid()
plt.xlabel('Time')
plt.legend()
plt.savefig('New_vs_Old_Evac_tight.png',dpi=300)
plt.show()
```

Same could be said for moving just responders

```python
fig = pyplot.figure(figsize=(5,5), dpi=80)
ax = fig.add_subplot(1,1,1)
plt.hold()
plt.hist(df['totalTime_resp_old'], 15, alpha=0.7, color='orange', label='Old Responce')
plt.hist(df['totalTime_new'], 30, alpha=0.7, color='b', label='New Total')
plt.xlim((10,40))
plt.grid()
plt.xlabel('Time')
plt.legend()
plt.savefig('New_vs_Old_Resp_tight.png',dpi=300)
plt.show()
```

However, we can see that it is clearly optimal to move both parties together in cooperation rather than one after another.  
Thus the simple advantage in using a multiparty route planner within capacity constrained environments.

```python
fig = pyplot.figure(figsize=(5,5), dpi=80)
ax = fig.add_subplot(1,1,1)
plt.hold()
plt.hist(df['totalTime_old'], 30, alpha=0.7, color='r', label='Old Total')
plt.hist(df['totalTime_new'], 30, alpha=0.7, color='b', label='New Total')
plt.xlim((15,50))
plt.grid()
plt.xlabel('Time')
plt.legend()
plt.savefig('New_vs_Old_tight.png',dpi=300)
plt.show()
```
